```python
from openai.embeddings_utils import cosine_similarity
from openai.embeddings_utils import get_embedding
from dotenv import load_dotenv
import utils
import pandas as pd
```

```python
openai = utils.init_OpenAI()
embeddings = utils.init_embeddings()
```

```python
#we use the tenacity library to create delays and retries when calling openAI to avoid hitting throtlling limits
def calc_embeddings(text):
    # replace newlines, which can negatively affect performance.
    text = text.replace("\n", " ")
    return get_embedding(text, engine=utils.OPENAI_ADA_EMBEDDING_DEPLOYMENT_NAME)
```

```python
datafile_path = "./data/arxiv/arxiv_articles.csv"
original_df = pd.read_csv(datafile_path)
```

```python
# split dataframe into dataframes of 1000 rows each
count = 0
dfs = [original_df[i:i+1000] for i in range(0,original_df.shape[0],1000)]
for df in dfs:
    newdf = df.copy()
    newdf["PaperTitle_embeddings"] = newdf.PaperTitle.apply(lambda x: calc_embeddings(x))
    newdf.to_csv(f"./data/arxiv/arxiv_articles_embeddings_{count}.csv")
    count += 1
    print(f"Saved {count} rows to file")
```
